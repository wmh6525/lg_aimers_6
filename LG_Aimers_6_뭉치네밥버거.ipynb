# 3. Making feature

```python
def making_features(df):
    more_feature_df = df.copy()
    epsilon = 1e-6  

    # 총 난자 수
    more_feature_df['총 난자 수'] =  more_feature_df[['미세주입된 난자 수', '수집된 신선 난자 수', '혼합된 난자 수']].sum(axis=1)
    more_feature_df['신선 난자 비중'] =  more_feature_df['수집된 신선 난자 수'] / ( more_feature_df['총 난자 수'] + epsilon)
    more_feature_df['혼합 난자 비중'] =  more_feature_df['혼합된 난자 수'] / ( more_feature_df['총 난자 수'] + epsilon)
    more_feature_df['혼합 난자 대비 기증자'] =  more_feature_df['혼합된 난자 수'] / ( more_feature_df['기증자 정자와 혼합된 난자 수'] + epsilon)
    more_feature_df['전체 배아 생성률'] = np.where(more_feature_df['총 난자 수'] > 0, more_feature_df['총 생성 배아 수'] / (more_feature_df['혼합된 난자 수'] + epsilon), 0)
    more_feature_df['총 난자 대비 저장된 배아 비율'] = np.where(more_feature_df['총 난자 수'] > 0, more_feature_df['저장된 배아 수'] / (more_feature_df['총 난자 수'] + epsilon), np.nan)


    # 신선 난자 활용
    more_feature_df['신선 난자 활용도'] =  more_feature_df['총 생성 배아 수'] / ( more_feature_df['수집된 신선 난자 수'] + epsilon)
    more_feature_df['신선 난자 활용도 x 총 난자'] =  more_feature_df['신선 난자 활용도'] *  more_feature_df['총 난자 수']
    more_feature_df['수집 대비 저장된 신선 난자 수'] = more_feature_df['수집된 신선 난자 수'] / (more_feature_df['저장된 신선 난자 수'] + epsilon)
    
    # 주입 성공률
    more_feature_df['주입 성공률'] =  more_feature_df['미세주입에서 생성된 배아 수'] / ( more_feature_df['미세주입된 난자 수'] + epsilon)
    more_feature_df['주입 성공률 x 총 난자'] =  more_feature_df['주입 성공률'] *  more_feature_df['총 난자 수']
    more_feature_df['주입 성공률 x 신선 활용도'] =  more_feature_df['주입 성공률'] *  more_feature_df['신선 난자 활용도']
    more_feature_df['주입 대비 신선 활용'] =  more_feature_df['주입 성공률'] / ( more_feature_df['신선 난자 활용도'] + epsilon)

    
    # 미세주입
    more_feature_df['미세주입 비중'] =  more_feature_df['미세주입된 난자 수'] / ( more_feature_df['총 난자 수'] + epsilon)
    more_feature_df['미세주입 저장률'] =  more_feature_df['미세주입 후 저장된 배아 수'] / ( more_feature_df['미세주입에서 생성된 배아 수'] + epsilon)

    # 배아 저장
    more_feature_df['배아 저장 비율'] =  more_feature_df['저장된 배아 수'] / ( more_feature_df['총 생성 배아 수'] + epsilon)
    more_feature_df['배아 생산율'] =  more_feature_df['총 생성 배아 수'] / ( more_feature_df['혼합된 난자 수'] + epsilon)
    more_feature_df['배아 저장 손실'] =  more_feature_df['총 생성 배아 수'] -  more_feature_df['저장된 배아 수']
    more_feature_df['배아 저장 비율 x 생산율'] =  more_feature_df['배아 저장 비율'] *  more_feature_df['배아 생산율']
    more_feature_df['배아 저장 비율 x 비생산율'] = np.where((more_feature_df['배아 저장 비율 x 생산율'] == 0) & ~more_feature_df['배아 저장 비율 x 생산율'].isna(), 0, 1)
    more_feature_df['전체 배아 생산율 대비 배아 저장 비율'] = more_feature_df['배아 저장 비율'] / (more_feature_df['배아 생산율'] + epsilon)
    

    # 배아 해동
    more_feature_df['해동 배아 활용도'] =  more_feature_df['해동된 배아 수'] / ( more_feature_df['저장된 배아 수'] + epsilon)   
    more_feature_df['배아 생존율'] =  more_feature_df['이식된 배아 수'] / ( more_feature_df['해동된 배아 수'] + epsilon)
    more_feature_df['배아 이식률'] =  more_feature_df['이식된 배아 수'] / ( more_feature_df['총 생성 배아 수'] + epsilon)

    # 시술
    more_feature_df['현재 시술용 포함 여부'] =  more_feature_df['배아 생성 주요 이유'].astype(str).str.contains('현재 시술용', na=True).astype(int)
    more_feature_df['이식 대비 경과일'] =  more_feature_df['이식된 배아 수'] / ( more_feature_df['배아 이식 경과일'] + epsilon)

    # 임신 시도
    more_feature_df['임신 도전'] =  more_feature_df['임신 시도 또는 마지막 임신 경과 연수'].notnull().astype(int)
    more_feature_df['이식 대비 저장'] =  more_feature_df['이식된 배아 수'] / ( more_feature_df['저장된 배아 수'] + epsilon)
    more_feature_df['이식 대비 혼합 난자'] =  more_feature_df['이식된 배아 수'] / ( more_feature_df['혼합된 난자 수'] + epsilon)
    more_feature_df['파트너 정자 비율'] = np.where(more_feature_df['혼합된 난자 수'] > 0, more_feature_df['파트너 정자와 혼합된 난자 수'] / (more_feature_df['혼합된 난자 수'] + epsilon), 0)

    # 불임 원인
    reason_1 = ["남성 주 불임 원인", "남성 부 불임 원인", "여성 주 불임 원인", "여성 부 불임 원인", "부부 주 불임 원인", "부부 부 불임 원인", "불명확 불임 원인"]
    more_feature_df['부부 불임 합'] = more_feature_df[reason_1].sum(axis=1)
    reason_2 = ["불임 원인 - 난관 질환", "불임 원인 - 남성 요인", "불임 원인 - 배란 장애", "불임 원인 - 여성 요인", "불임 원인 - 자궁경부 문제", "불임 원인 - 자궁내막증", "불임 원인 - 정자 농도", "불임 원인 - 정자 면역학적 요인", "불임 원인 - 정자 운동성", "불임 원인 - 정자 형태"]
    more_feature_df['불임 세부 합'] = more_feature_df[reason_2].sum(axis=1)

    
    
    return  more_feature_df
```

```python
train_df = making_features(train_df)
test_df = making_features(test_df)
```

```python
def get_catboost_params(trial):
    return {
        "eval_metric": "AUC",
        "verbose": 0,
        'iterations': 1200,
        'od_wait': trial.suggest_int('od_wait', 500, 2300),
        'learning_rate': trial.suggest_uniform('learning_rate', 0.001, 1),
        'reg_lambda': trial.suggest_uniform('reg_lambda', 1e-5, 100),
        'random_strength': trial.suggest_uniform('random_strength', 10, 50),
        'depth': trial.suggest_int('depth', 1, 15),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 1, 30),
        'leaf_estimation_iterations': trial.suggest_int('leaf_estimation_iterations', 1, 15),
        'bagging_temperature': trial.suggest_loguniform('bagging_temperature', 0.01, 100.00),
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.4, 1.0),
        'scale_pos_weight': trial.suggest_int('scale_pos_weight', 1, 3)
    }

def train_and_evaluate(X_train, X_valid, y_train, y_valid, params, cat_features):
    train_pool = Pool(X_train, y_train, cat_features=cat_features)
    eval_pool = Pool(X_valid, y_valid, cat_features=cat_features)

    model = CatBoostClassifier(**params)
    model.fit(train_pool, eval_set=eval_pool, use_best_model=True, early_stopping_rounds=50)

    preds_valid = model.predict_proba(eval_pool)[:, 1]
    auc = roc_auc_score(y_valid, preds_valid)

    return auc

def optimize_catboost(trial):
    catboost_params = get_catboost_params(trial)
    
    kf = KFold(n_splits=5, shuffle=True, random_state=21)
    auc_scores = []

    for fold, (train_idx, valid_idx) in enumerate(kf.split(X_encoded, y)):
        print(f"{'=' * 10} Fold {fold+1} {'=' * 10}")

        X_train, X_valid = X_encoded.iloc[train_idx], X_encoded.iloc[valid_idx]
        y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

        auc = train_and_evaluate(X_train, X_valid, y_train, y_valid, catboost_params, cat_features)
        auc_scores.append(auc)

        print(f"Fold {fold+1}, AUC: {auc:.4f}")

    mean_auc = np.mean(auc_scores)
    print(f"Mean AUC for this trial: {mean_auc:.4f}")

    return mean_auc

sampler = TPESampler(seed=42)

catboost_optimization = optuna.create_study(direction='maximize', sampler=sampler)
catboost_optimization.optimize(optimize_catboost, n_trials=100)

print("Best parameters:", catboost_optimization.best_params)
print("Best AUC:", catboost_optimization.best_value)
```

```python
lgbm_predictions = []
lgbm_feature_importance = []
lgbm_scores = []

def get_lgbm_params(trial):
    return {
        'num_leaves': trial.suggest_int('num_leaves', 2, 1024, log=True), 
        'max_depth': trial.suggest_int('max_depth', 1, 10), 
        'learning_rate': trial.suggest_float('learning_rate', 0.0001, 0.1, log=True), 
        'n_estimators': trial.suggest_int('n_estimators', 8, 1024, log=True), 
        'objective': 'binary', 
        'class_weight': trial.suggest_categorical('class_weight', ['balanced', None]),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 50), 
        'subsample': trial.suggest_float('subsample', 0.7, 1.0), 
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.7, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 1.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 10.0),
        'random_state': 42,
        'metric': 'auc',
        'verbosity': 0
    }

def train_and_evaluate(X_train, X_valid, y_train, y_valid, params):
    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train)

    preds_valid = model.predict_proba(X_valid)[:, 1]
    auc = roc_auc_score(y_valid, preds_valid)

    return auc

def optimize_lgbm(trial):
    lgbm_params = get_lgbm_params(trial)
    
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    auc_scores = []

    for fold, (train_idx, valid_idx) in enumerate(kf.split(X_encoded, y)):
        print(f"{'=' * 10} Fold {fold+1} {'=' * 10}")

        X_train, X_valid = X_encoded.iloc[train_idx], X_encoded.iloc[valid_idx]
        y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

        auc = train_and_evaluate(X_train, X_valid, y_train, y_valid, lgbm_params)
        auc_scores.append(auc)

        print(f"Fold {fold+1}, AUC: {auc:.4f}")

    mean_auc = np.mean(auc_scores)
    print(f"Mean AUC for this trial: {mean_auc:.4f}")

    return mean_auc

sampler = TPESampler(seed=42)

lgbm_optimization = optuna.create_study(direction='maximize', sampler=sampler)
lgbm_optimization.optimize(optimize_lgbm, n_trials=50)

print("Best parameters:", lgbm_optimization.best_params)
print("Best AUC:", lgbm_optimization.best_value)
```
